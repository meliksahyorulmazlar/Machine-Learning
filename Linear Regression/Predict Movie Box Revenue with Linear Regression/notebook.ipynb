```python
# Predict Movie Box Revenue with Linear Regression
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("cost_revenue_clean.csv")
data.describe()
```

```python
data.shape
```

```python
x = data[['production_budget_usd']]
y = data[['worldwide_gross_usd']]
```

```python

plt.figure(figsize=(15,10))
plt.title("Using Movie Budgets to predict Movie Box Revenue")
plt.ylabel("Movie Revenue ($)")
plt.xlabel("Movie Budget ($)")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(0,450_000_000)
plt.ylim(0,3_000_000_000)
plt.scatter(x,y,alpha=0.3)
plt.show()
```

```python
# Linear Regression
regression = LinearRegression()
regression.fit(x,y)
print(f"Theta one is {regression.coef_[0][0]},also known as the gradient.")
print(f"Theta zero is {regression.intercept_[0]},also known the y-intercept.")
```

```python
plt.figure(figsize=(15,10))
plt.title("Using Movie Budgets to predict Movie Box Revenue")
plt.ylabel("Movie Revenue ($)")
plt.xlabel("Movie Budget ($)")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(0,450_000_000)
plt.ylim(0,3_000_000_000)
plt.scatter(x,y,alpha=0.3)
plt.plot(x,regression.predict(x),color='r',linewidth=3)
plt.show()
```

```python
# The following code will return the variance in the target by the model's single target
regression.score(x,y)
```

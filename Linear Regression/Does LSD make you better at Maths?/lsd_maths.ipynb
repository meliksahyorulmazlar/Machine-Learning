```python
# Does LSD usage make you better at maths?
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("lsd_math_score_data.csv")
data.describe()
```

```python
data.shape
```

```python
x = data[['LSD_ppm']]
y = data[['Avg_Math_Test_Score']]
```

```python
plt.title("Does LSD make you better at Maths?")
plt.ylabel("Math Score out of 100")
plt.xlabel("LSD PPM")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.scatter(x,y)
plt.show()
```

```python
# Linear Regression
regression = LinearRegression()
regression.fit(x,y)
print(f"Theta one is {regression.coef_[0][0]},also known as the gradient.")
print(f"Theta zero is {regression.intercept_[0]},also known the y-intercept.")
```

```python
plt.title("Does LSD make you better at Maths?")
plt.ylabel("Math Score out of 100")
plt.xlabel("LSD PPM")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.scatter(x,y)
plt.plot(x,regression.predict(x),color='red',linewidth=3)
plt.show()
```
